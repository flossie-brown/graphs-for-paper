```python
import iris 
import matplotlib.pyplot as plt
import iris.quickplot as qplt
import iris.plot as iplt
import numpy as np

%matplotlib inline
```

```python
data_dir = '/shared/netscratch/lrm49/UKCA_VN112_QBO_SIMS/'
```

```python
JAN_W = iris.load(data_dir+'JAN_W/u-bl775_all_vars_processed.nc')
a = iris.load(data_dir+'JAN_W/u-bo186_all_vars_processed.nc')
b = iris.load(data_dir+'JAN_W/u-bo189_all_vars_processed.nc')
c= iris.load(data_dir+'JAN_W/u-bo190_all_vars_processed.nc')
JUL_E= iris.load(data_dir+'JULY_E/u-bl775_all_vars_processed.nc')
j = iris.load(data_dir+'JULY_E/u-bo005_all_vars_processed.nc')
k = iris.load(data_dir+'JULY_E/u-bo053_all_vars_processed.nc')
l = iris.load(data_dir+'JULY_E/u-bo056_all_vars_processed.nc')
JUL_W = iris.load(data_dir+'JULY_W/u-bl775_all_vars_processed.nc')
d = iris.load(data_dir+'JULY_W/u-bn515_all_vars_processed.nc')
e = iris.load(data_dir+'JULY_W/u-bn941_all_vars_processed.nc')
f = iris.load(data_dir+'JULY_W/u-bn942_all_vars_processed.nc')
JAN_E = iris.load(data_dir+'JAN_E/u-bl775_all_vars_processed.nc')
g = iris.load(data_dir+'JAN_E/u-bo332_all_vars_processed.nc')
h = iris.load(data_dir+'JAN_E/u-bo334_all_vars_processed.nc')
i = iris.load(data_dir+'JAN_E/u-bo335_all_vars_processed.nc')
```

```python
master_list = [a,b,c,d,e,f,g,h,i,j,k,l]
```

```python
control_list = [JAN_W, JUL_W, JAN_E, JUL_E]
```

## Calculate temp anomaly

```python
import iris.analysis.cartography
```

```python
temp_list = []
for cube in master_list:
        temp = cube[5]
        temp_list.append(temp)
```

```python
temp_list_c = []
for cube in control_list:
    temp = cube[5]
    temp_list_c.append(temp)
```

## Subtract the correct anomaly

```python
temp_anomaly = []
for i in range(0, 4): #there are 4 controls
    for eruption in temp_list[i*3: (i*3)+3]: #split into 4 chunks of 3
        anomaly = eruption - temp_list_c[i] #subtract the correct control from each chunk
        temp_anomaly.append(anomaly)
```

## Constrain latitude to tropics

```python
temp_tropics = []
for eruption in temp_anomaly:
    tropics = eruption.extract(iris.Constraint(latitude = lambda cell: -30<cell<30))
    temp_tropics.append(tropics)
```

### Guess bounds

```python
temp_tropics[0].coord('latitude').guess_bounds()
temp_tropics[0].coord('longitude').guess_bounds()
```

```python
grid_areas_trop = iris.analysis.cartography.area_weights(temp_tropics[0])
```

## Collapse over lat and lon

```python
temp_collapsed = []
for eruption in temp_tropics[0:4]:
    trop = eruption.collapsed(['latitude', 'longitude'], iris.analysis.MEAN, weights = grid_areas_trop)
    temp_collapsed.append(trop)
```

```python
for eruption in temp_tropics[4:8]:
    trop = eruption.collapsed(['latitude', 'longitude'], iris.analysis.MEAN, weights = grid_areas_trop)
    temp_collapsed.append(trop)
```

```python
for eruption in temp_tropics[8:12]:
    trop = eruption.collapsed(['latitude', 'longitude'], iris.analysis.MEAN, weights = grid_areas_trop)
    temp_collapsed.append(trop)
```

# Repeat for SO4

```python
SO4_list = []
for cube in master_list:
    SO4 = cube[2]
    SO4_list.append(SO4)
```

```python
SO4_list_c = []
for cube in control_list:
    SO4 = cube[2]
    SO4_list_c.append(SO4)
```

```python
SO4_anomaly = []
for i in range(0, 4):
    for eruption in SO4_list[i*3: (i*3)+3]:
        anomaly = eruption - SO4_list_c[i]
        SO4_anomaly.append(anomaly)
```

```python
SO4_tropics = []
for eruption in SO4_anomaly:
    trop = eruption.extract(iris.Constraint(latitude = lambda cell: -30<cell<30))
    SO4_tropics.append(trop)
```

```python
SO4_collapsed = []
for eruption in SO4_tropics[0:4]:
    trop = eruption.collapsed(['latitude', 'longitude'], iris.analysis.MEAN, weights = grid_areas_trop)
    SO4_collapsed.append(trop)
```

```python
for eruption in SO4_tropics[4:8]:
    trop = eruption.collapsed(['latitude', 'longitude'], iris.analysis.MEAN, weights = grid_areas_trop)
    SO4_collapsed.append(trop)
```

```python
for eruption in SO4_tropics[8:12]:
    trop = eruption.collapsed(['latitude', 'longitude'], iris.analysis.MEAN, weights = grid_areas_trop)
    SO4_collapsed.append(trop)
```

## Set up plot

```python
SO4_level = ( 2e-8, 10e-8)
```

```python
# Convert m to km
labels = range(10,50,5) # create labels
positions = range(10000,50000,5000) # identify positions for labels
```

## Find pressures associated with altitudes

```python
pressure = JUL_W[4] #select cube of pressure levels on altitude levels
```

```python
tropics_p = pressure.extract(iris.Constraint(latitude = lambda cell: -30<cell<30))
collapsed_p=tropics_p.collapsed(['latitude', 'longitude'], iris.analysis.MEAN, weights = grid_areas_trop) 
time_collapsed_p = collapsed_p.collapsed('time', iris.analysis.MEAN) #find average pressure over altitudes used
```

```python
altitude_m = [40000,30950,16550,10000] # select altitudes to convert to pressure
```

```python
pressure_pa = np.interp(altitude_m, SO4_collapsed[0].coord('altitude').points, time_collapsed_p.data) #interpolate to find pressures associated with altitude
```

```python
pressure_values = [int(x/100) for x in pressure_pa] # convert to hPa and to integer values
```

```python
altitude_values = [x - 10000 for x in altitude_m] #convert to km and realign to match xlim values
```

```python
print(pressure_values, altitude_values)
```

```python
# create a pressure axis
pressure_labels = pressure_values
pressure_positions = altitude_values
```

```python
temp_levels = np.linspace(-10,10,11)
titles = ['January, w-QBO:', 'July, w-QBO:', 'January, e-QBO:', 'July, e-QBO:']
letter_labels = ('a)', 'b)', 'c)', 'd)', 'e)', 'f)', 'g)', 'h)', 'i)','j)' ,'k)' ,'l)' )
names = ['EQ', '45$^\circ$N', '45$^\circ$S ', 'EQ', '45$^\circ$N ', '45$^\circ$S ', 'EQ', '45$^\circ$N', '45$^\circ$S', 'EQ', '45$^\circ$N ', '45$^\circ$S']
x = range(0,36)
```

```python
fig, axs = plt.subplots(4,3,sharey = True, sharex = True, figsize = (17,27))

ax = axs.ravel()

for i, eruption in enumerate(temp_collapsed):
    y = eruption.coord('altitude').points
    z = np.transpose(eruption.data)
    cf = ax[i].contourf(x,y,z,temp_levels, extend = 'both', cmap = plt.cm.RdYlBu_r)
    ax[i].set_ylim(10000,40000)
    ax[i].xaxis.set_tick_params(width=2, labelsize = 25, direction = 'out', length = 6)
    ax[i].yaxis.set_tick_params(width=2, labelsize = 25, direction = 'out', length = 6)
    ax[i].set_title(letter_labels[i] +'   ' + names[i], size=25, loc = 'left', y = 1.05)
  
for i, eruption in enumerate(SO4_collapsed):
    z = np.transpose(eruption.data)
    ax[i].contour(x,y,z,SO4_level, cmap = plt.cm.Dark2, linewidths = 5)
    
for i in range(0, 10,3):
    ax1= ax[i+2].twinx()
    ax1.set_yticks(pressure_positions)
    ax1.set_yticklabels(pressure_labels)
    ax1.set_ylabel('Pressure / hPa', fontsize = 20)
    ax1.tick_params(labelsize = 15)
    ax[i].set_yticklabels(labels) 
    ax[i].set_ylabel('Altitude / km', fontsize = 30, labelpad = 18) 
    ax[i+1].set_title(titles[i/3], size=28, y = 1.19, fontweight = 'bold')
    
    
for i in range(9,12):
    ax[i].set_xlabel('Months since eruption', fontsize = 30, labelpad = 25)    
    
  
cbar_ax = fig.add_axes([0.1, 0.1, 0.8, 0.02])
colourbar = plt.colorbar(cf, cax = cbar_ax , orientation = 'horizontal')
colourbar.set_label('K', fontsize = 35)
colourbar.ax.tick_params(labelsize=30)

plt.tight_layout() 
plt.subplots_adjust( bottom = 0.2, wspace = 0.1, hspace = 0.35) 
plt.savefig('60Tg_eruptions.png')
```
