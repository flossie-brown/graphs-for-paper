```python
import iris 
import matplotlib.pyplot as plt
import iris.quickplot as qplt
import iris.plot as iplt
import numpy as np

%matplotlib inline
```

```python
data_dir = '/shared/netscratch/lrm49/UKCA_VN112_QBO_SIMS/'
```

```python
JAN_W = iris.load(data_dir+'JAN_W/u-bl775_all_vars_processed.nc')
a = iris.load(data_dir+'JAN_W/u-bo186_all_vars_processed.nc')
b = iris.load(data_dir+'JAN_W/u-bo189_all_vars_processed.nc')
c= iris.load(data_dir+'JAN_W/u-bo190_all_vars_processed.nc')
JUL_E= iris.load(data_dir+'JULY_E/u-bl775_all_vars_processed.nc')
j = iris.load(data_dir+'JULY_E/u-bo005_all_vars_processed.nc')
k = iris.load(data_dir+'JULY_E/u-bo053_all_vars_processed.nc')
l = iris.load(data_dir+'JULY_E/u-bo056_all_vars_processed.nc')
JUL_W = iris.load(data_dir+'JULY_W/u-bl775_all_vars_processed.nc')
d = iris.load(data_dir+'JULY_W/u-bn515_all_vars_processed.nc')
e = iris.load(data_dir+'JULY_W/u-bn941_all_vars_processed.nc')
f = iris.load(data_dir+'JULY_W/u-bn942_all_vars_processed.nc')
JAN_E = iris.load(data_dir+'JAN_E/u-bl775_all_vars_processed.nc')
g = iris.load(data_dir+'JAN_E/u-bo332_all_vars_processed.nc')
h = iris.load(data_dir+'JAN_E/u-bo334_all_vars_processed.nc')
i = iris.load(data_dir+'JAN_E/u-bo335_all_vars_processed.nc')
```

```python
master_list = [a,b,c,d,e,f,g,h,i,j,k,l]
```

```python
control_list = [JAN_W, JUL_W, JAN_E, JUL_E]
```

## Select SAOD cube

```python
import iris.analysis.cartography
```

```python
SAOD_list = []
for cube in master_list:
    SAOD = cube[0]
    SAOD_list.append(SAOD)
```

```python
SAOD_list_c = []
for cube in control_list:
    SAOD = cube[0]
    SAOD_list_c.append(SAOD)
```

## Subtract the correct anomaly

```python
SAOD_anomaly = []
for i in range(0, 4): #there are 4 controls
    for eruption in SAOD_list[i*3: (i*3)+3]: #split into 4 chunks of 3
        anomaly = eruption - SAOD_list_c[i] #subtract the correct control from each chunk
        SAOD_anomaly.append(anomaly)
```

## Constrain latitude to tropics

```python
zonal_SAOD = []
for eruption in SAOD_list:
    lon_collapsed = eruption.collapsed('longitude', iris.analysis.MEAN)
    zonal_SAOD.append(lon_collapsed)
```

## Set up plot

```python
SAOD_levels = np.linspace(0, 2, 11)
```

```python

titles = ['January, w-QBO:', 'July, w-QBO:', 'January, e-QBO:', 'July, e-QBO:']
letter_labels = ('a)', 'b)', 'c)', 'd)', 'e)', 'f)', 'g)', 'h)', 'i)','j)' ,'k)' ,'l)' )
names = ['EQ', '45$^\circ$N', '45$^\circ$S ', 'EQ', '45$^\circ$N ', '45$^\circ$S ', 'EQ', '45$^\circ$N', '45$^\circ$S', 'EQ', '45$^\circ$N ', '45$^\circ$S']
x = range(0,36)
```

```python
fig, axs = plt.subplots(4,3,sharey = True, sharex = True, figsize = (17,27))

ax = axs.ravel()

for i, eruption in enumerate(zonal_SAOD):
    y = eruption.coord('latitude').points
    z = np.transpose(eruption.data)
    cf = ax[i].contourf(x,y,z,SAOD_levels, extend = 'both', cmap = plt.cm.plasma)
    ax[i].xaxis.set_tick_params(width=2, labelsize = 25, direction = 'out', length = 6)
    ax[i].yaxis.set_tick_params(width=2, labelsize = 25, direction = 'out', length = 6)
    ax[i].set_title(letter_labels[i] +'   ' + names[i], size=25, loc = 'left', y = 1.05)
  
    
for i in range(0, 10,3):
    ax[i].set_ylabel('latitude / degreees', fontsize = 30, labelpad = 18) 
    ax[i+1].set_title(titles[i/3], size=28, y = 1.19, fontweight = 'bold')
    
    
for i in range(9,12):
    ax[i].set_xlabel('Months since eruption', fontsize = 30, labelpad = 25)    
    
  
cbar_ax = fig.add_axes([0.1, 0.1, 0.8, 0.02])
colourbar = plt.colorbar(cf, cax = cbar_ax , orientation = 'horizontal')
colourbar.ax.tick_params(labelsize=30)

plt.tight_layout() 
plt.subplots_adjust( bottom = 0.2, wspace = 0.1, hspace = 0.35) 
plt.savefig('SAOD_60Tg_eruptions.png')
```
