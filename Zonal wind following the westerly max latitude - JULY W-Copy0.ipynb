### Average zonal wind over longitude

```python
list_x = []
for cube in master_list:
    x = cube[11]
    list_x.append(x)
```

```python
control_list = [control_master[10], JNW1[11],JNW2[11]]
```

```python
x_average = []
for eruption in list_x:
    lon_average = eruption.collapsed( 'longitude', iris.analysis.MEAN)
    x_average.append(lon_average)
```

```python
cont_average = []
for eruption in control_list:
    lon_average = eruption.collapsed( 'longitude', iris.analysis.MEAN)
    cont_average.append(lon_average)
```

# A: Zonal wind at equator

```python
latlist = []
for eruption in x_average:
    lat_constraint = eruption.extract(iris.Constraint(latitude = lambda cell: -2 <= cell <=2))
    latlist.append(lat_constraint)
```

```python
areaav_list = []
for eruption in latlist:
    area_average = eruption.collapsed('latitude', iris.analysis.MEAN)
    areaav_list.append(area_average)
```

```python
latlist_cont = []
for eruption in control_list:
    lat_constraint = eruption.extract(iris.Constraint(latitude = lambda cell: -2 <= cell <=2))
    latlist_cont.append(lat_constraint)
```

```python
areaav_list_cont = []
for eruption in latlist_cont:
    area_average = eruption.collapsed(['latitude', 'longitude'], iris.analysis.MEAN)
    areaav_list_cont.append(area_average)
```

### Extract tropics

```python
x_tropics = []
for eruption in x_average:
    lat = eruption.extract(iris.Constraint(latitude = lambda cell: -30<cell<30))
    press = lat.extract(iris.Constraint(pressure = lambda cell: 10<cell<100))
    x_tropics.append(press)
```

```python
cont_tropics = []
for eruption in cont_average:
    lat = eruption.extract(iris.Constraint(latitude = lambda cell: -30<cell<30))
    press = lat.extract(iris.Constraint(pressure = lambda cell: 10<cell<100))
    cont_tropics.append(press)
```

# B: Zonal wind at latitudinal position of QBO

### Find maximum x wind at each latitude point

```python
zonal_maxima = []
for eruption in x_tropics:
    result = eruption.collapsed('pressure', iris.analysis.MAX)
    zonal_maxima.append(result)
```

```python
control_maxima = []
for eruption in cont_tropics:
    result= eruption.collapsed('pressure', iris.analysis.MAX)
    control_maxima.append(result)
```

### Find the absolute maxium value at each time slice

```python
max_list_list = []
for eruption in zonal_maxima:
    max_list = []
    for slices in eruption.slices_over('time'):
        constrained_lat = slices.extract(iris.Constraint(latitude = lambda cell: -15<cell<15))
        max_x = max(constrained_lat.data)
        max_list.append(max_x)
    max_list_list.append(max_list)
```

```python
max_list_cont = []
for eruption in control_maxima:
    max_cont = []
    for slices in eruption.slices_over('time'):
        constrained_lat = slices.extract(iris.Constraint(latitude = lambda cell: -15<cell<15))
        max_x = max(constrained_lat.data)
        max_cont.append(max_x)
    max_list_cont.append(max_cont)
```

### Find latitude value associated with maxima

```python
lat_list_list = []
for b, eruption in enumerate(zonal_maxima):
    lat_list = []
    for i, slices in enumerate(eruption.slices_over('time')):
        data =slices.data
        result = np.where(data == max_list_list[b][i])
        latitude = slices.coord('latitude').points[result]
        lat_list.append(latitude[0])
    lat_list_list.append(lat_list)
```

```python
lat_list_cont = []
for b, eruption in enumerate(control_maxima):
    lat_cont = []
    for i, slices in enumerate(eruption.slices_over('time')):
        data =slices.data
        result = np.where(data == max_list_cont[b][i])
        latitude = slices.coord('latitude').points[result]
        lat_cont.append(latitude[0])
    lat_list_cont.append(lat_cont)
```

### Constrain data by max latitude

```python
full_list = []
for b, eruption in enumerate(x_average):
    pressure_series = []
    for i, slices in enumerate(eruption.slices_over('time')):
        lat = lat_list_list[b][i]
        result = slices.extract(iris.Constraint(latitude = lat))
        pressure_series.append(result)
    full_list.append(pressure_series)
```

```python
cont_list = []
for b, eruption in enumerate(cont_average):
    pressure_series = []
    for i, slices in enumerate(eruption.slices_over('time')):
        lat = lat_list_cont[b][i]
        result = slices.extract(iris.Constraint(latitude = lat))
        pressure_series.append(result)
    cont_list.append(pressure_series)
```

### Make data back into an array

```python
plot_list = []
for eruption in full_list:
    plot_array = []
    for press in eruption:
        data = press.data
        plot_array.append(data)
    plot_list.append(plot_array)
```

```python
plot_list_cont = []
for eruption in cont_list:
    plot_array = []
    for press in eruption:
        data = press.data
        plot_array.append(data)
    plot_list_cont.append(plot_array)
```
