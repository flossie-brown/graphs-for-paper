# Constrain data to get equatorial x wind

### Select x wind

```python
list_x = [JNW15, JNW60, JLW15, JLW60, JNE15, JNE60, JLE15, JLE60]
```

```python
control_list = [JAN_W, JNW1,JNW2,  JUL_W, JLW1,JLW2, JAN_E, JNE1,JNE2, JUL_E, JLE1, JLE2] 
```

### Collapse over latitude

```python
latlist = []
for eruption in list(list_x):
    lat_constraint = eruption.extract(iris.Constraint(latitude = lambda cell: -2 <= cell <=2))
    latlist.append(lat_constraint)
```

```python
latlist_cont = []
for eruption in control_list:
    lat_constraint = eruption.extract(iris.Constraint(latitude = lambda cell: -2 <= cell <=2))
    latlist_cont.append(lat_constraint)
```

### Find bounds

```python
import iris.analysis.cartography
```

```python
latlist[0][0].coord('latitude').guess_bounds()
latlist[0][0].coord('longitude').guess_bounds()
```

```python
grid_areas = iris.analysis.cartography.area_weights(latlist[0][0])
```

### Average over area

```python
areaav_list = []
for eruption in latlist:
    area_average = eruption[0].collapsed(['latitude', 'longitude'], iris.analysis.MEAN, weights = grid_areas)
    areaav_list.append(area_average)
```

```python
areaav_list_cont = []
for eruptions in latlist_cont:
    for eruption in eruptions:
        area_average = eruption.collapsed(['latitude', 'longitude'], iris.analysis.MEAN)
        areaav_list_cont.append(area_average)
```

```python
control_list = [areaav_list_cont[0:3], areaav_list_cont[3:6], areaav_list_cont[6:9], areaav_list_cont[9:12]]
```

# Design plot

```python
zero_contour = 0
titles = ['January, w-QBO:', 'July, w-QBO:', 'January, e-QBO:', 'July, e-QBO:']
letter_labels = ('a)', 'b)', 'c)', 'd)', 'e)', 'f)', 'g)', 'h)', 'i)','j)' ,'k)' ,'l)' )
```

## Plot for x wind

```python
fig, axs = plt.subplots(4,2,sharey = True, sharex = True, figsize = (17,27))

ax = axs.ravel()

for i, eruption in enumerate(areaav_list):
        x = range(0, 36)
        y = eruption.coord('pressure').points
        z = np.transpose(eruption.data)
        ax[i].contour(x,y,z,zero_contour, linewidths = 10, colors = 'blue')
        ax[i].contourf(x,y,z,zero_contour, extend = 'max', alpha = 0.4)
        ax[i].set_yscale('log')
        
    
for i, season in enumerate(control_list):
    for b, eruption in enumerate(season):
        x = range(0, eruption.coord('time').shape[0])
        y = eruption[0].coord('pressure').points
        z = np.transpose(eruption.data)
        
        if b == 0:
            ax[2*i].contour(x,y,z,zero_contour,colors = 'black', linestyles = 'dashed', linewidths = 10)
            ax[2*i].contourf(x,y,z,zero_contour, extend = 'max', alpha = 0.1)
        else:
            ax[2*i].contour(x,y,z,zero_contour, colors = 'black', linestyles = 'dashed', linewidths = 4)
            ax[2*i].contourf(x,y,z,zero_contour, extend = 'max', alpha = 0.03)

    ax[2*i].set_title(titles[i],loc = 'right', fontsize = 30, y = 1.08, x = 1.25, fontweight = 'bold')
    ax[2*i].xaxis.set_tick_params(width = 1.5, labelsize = 25, direction = 'out', length = 5)
    ax[2*i].yaxis.set_tick_params(which = 'both', width = 1.5, labelsize = 25, direction = 'out', length = 5)
    ax[2*i].set_title(letter_labels[2*i] + '  15 Tg', size=25,  y = 1.03)
    ax[2*i].set_ylabel('Pressure / hPa', fontsize = 30, labelpad = 18)
    ax[2*i].set_ylim(5,100)
    ax[2*i].invert_yaxis()
    
for i, season in enumerate(control_list):
    num = (2*i)+1
    for b, eruption in enumerate(season):
        x = range(0, eruption.coord('time').shape[0])
        y = eruption[0].coord('pressure').points
        z = np.transpose(eruption.data)
        
        if b == 0:
            ax[num].contour(x,y,z,zero_contour,colors = 'black', linestyles = 'dashed', linewidths = 10)
            ax[num].contourf(x,y,z,zero_contour, extend = 'max', alpha = 0.1)
        else:
            ax[num].contour(x,y,z,zero_contour, colors = 'black', linestyles = 'dashed', linewidths = 4)
            ax[num].contourf(x,y,z,zero_contour, extend = 'max', alpha = 0.03)

    ax[num].xaxis.set_tick_params(width = 1.5, labelsize = 25, direction = 'out', length = 5)
    ax[num].yaxis.set_tick_params(which = 'both', width = 1.5, labelsize = 25, direction = 'out', length = 5)
    ax[num].set_title(letter_labels[num] + '  60 Tg', size=25, y = 1.03)
    ax[num].invert_yaxis()
    
ax[6].set_xlabel('Months since start of simulation', fontsize = 30, labelpad = 20)    
ax[7].set_xlabel('Months since start of simulation', fontsize = 30, labelpad = 20)


plt.tight_layout() 
plt.subplots_adjust(wspace = 0.1, hspace = 0.20) 
plt.savefig('2tone_60tg_eruptions.png')
```
